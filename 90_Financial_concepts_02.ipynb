## Simple vs compound interest

```python
def simple_intr(amount, r, period):
    return amount * r * period

def compound_intr(amount, r, period):
    return (amount * (1 + r)**period) - amount
```

```python
simple_intr(500000, .05, 3), compound_intr(500000, .05, 3)
```

## Compound doubling

```python
# How much time to double your money with compound?
def get_double_time(r):
    """
    Takes 
    - r <float>: compound interest
    
    Returns
    Time to double the investment
    """    
    return np.log(2) / np.log(1 + r)

# A simplified version of this function is called `rule 72`. It gives an approximation of doubling time
def rule72(r):
    """
    Takes 
    - r <float>: compound interest
    
    Returns
    An approximation of time to double the investment
    """    
    return 72 / (100 * r)
```

This function can model other exponentially growing phenomena, like GDP.

```python
# Let's assume that GDP grow 4% annualy
# How long will it take to double the economy?
get_double_time(.04)
```

Other iteresting contexts would be:

* A mutual fund that charges 3% in annual expense fees will reduce the investment principal to half in around 24 years. 


* A borrower who pays 12% interest on his credit card (or any other form of loans which is charging compound interest) will double the amount he owes in six years.


* If inflation is 6%, then a given purchasing power of the money will be worth half in around (72 รท 6) = 12 years. If inflation decreases from 6% to 4%, an investment will be expected to lose half its value in 18 years, instead of 12 years.

Source: https://www.investopedia.com/terms/r/ruleof72.asp?utm_source=investing-basics-new&utm_campaign=www.investopedia.com&utm_term=&utm_medium=email


```python
# Let's assume I have $10000 debt on my credit card and the interest is 10%
get_double_time(.1)
```

```python
# In seven years my debt will be doubled.
# What does the heuristic say?
rule72(.1)
```

```python
# Let's compare both
rates = np.arange(0.05, 1.05, .05)
```

```python
plot_series([pd.Series(get_double_time(rates), index = rates), pd.Series(rule72(rates), index = rates)], 
            ['Exponential', 'Approx.'], 
            'Compound doubling', lw = 5)
```
