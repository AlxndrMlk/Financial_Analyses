## Annuity example

You want to retire and maintain a monthly income of $2500 for the next 20 years. 

How much would it cost to purchase this annuity if discout rates are currently 4\%?

```python
# Assumptions
yr_disc_rt = .04
mo_inc = 2500
period_yr = 20

# Compute necessary quantities
period_mo = period_yr * 12
mo_disc_rt = (1 + yr_disc_rt)**(1/12) - 1
```

```python
# Sanity check
total = 1

for i in range(12):
    total += total * mo_disc_rt 
    
total, 1 * 1.04 # Yearly 4%
```

Get present values of cashflows:

```python
present_vals = [mo_inc / (1 + mo_disc_rt)**(i + 1) for i in range(period_mo)]
```

```python
plt.plot(present_vals)
plt.show()
```

```python
sum(present_vals), sum(present_vals) / 240
```

## Annuity example 2

If an annuity pays $500 monthly for 10 yrs at 15\% annual interst rate compounded monthly, what is the annuity worth today?

```python
# Assumptions
mo_payment = 500
periods_yr = 10
yr_intr_rt = .15

# Compute necessary quantities
periods_mo = periods_yr * 12
mo_intr_rt = (1 + yr_intr_rt)**(1/12) - 1
```

```python
mo_intr_rt
```

```python
vals = []

for i in range(periods_mo):
    vals.append(mo_payment / (1 + mo_intr_rt)**(i + 1))
```

```python
sum(vals)
```

```python
plt.plot(vals)
plt.show()
```
