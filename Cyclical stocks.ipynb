```python
import numpy as np
import pandas_datareader.data as web

from datetime import datetime as dt

import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

# Cyclical stocks

Stocks like **Caterpillar** or **Komatsu** follow the market cyclicity. Let's see this.

```python
# Define date range
start = dt(1980,  1, 2)
end   = dt(2020, 11, 20)
```

```python
# Get stock prices for FB
cat   = web.DataReader('CAT', 'stooq', start, end).sort_index()
```

```python
plt.figure(figsize = (12, 5))
plt.plot(cat.Close, label = 'Caterpillar', lw = 1)
plt.legend()
plt.title('Closing prices')
plt.plot()
```
